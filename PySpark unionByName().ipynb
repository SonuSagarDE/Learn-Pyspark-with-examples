#####PySpark unionByName() is used to union two DataFrames when you have column names in a different order or even if you have missing columns in any DataFrame

```
####unionByName() Syntax
unionByName(df, allowMissingColumns=True)
```

```
df1=spark.createDataFrame([[5,6,7]],["col0","col1","col2"])
df2=spark.createDataFrame([[6,8,9]],["col1","col4","col3"])

df_union=df1.unionByName(df2,allowMissingColumns=True)
df_union.show()
```
