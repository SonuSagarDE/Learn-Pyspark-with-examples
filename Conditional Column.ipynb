```
## Scenerios :: Create a Flag as Y and N , check for latest week which is having value more than zero will be considered as Y and else will be N
```

```
sample_data=[('2022-41',20,),\
                         ('2022-42',30,),\
                         ('2022-43',40,),\
                         ('2022-44',20,),\
                         ('2022-45',0,),\
                         ('2022-46',0,),\
                         ('2022-48',0,),
                        ]
columns=["Week","Value"]
df=spark.createDataFrame(data=sample_data,schema=columns)
df_1=df.filter("Value>0")
df_1.show()
```

```
from pyspark.sql import functions as F
latest_week=F.max("Week").alias("latest_week")
df_2=df_1.select(latest_week)
df_2.show()
```
