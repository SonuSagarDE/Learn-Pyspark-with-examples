```
SampleData_1=[(1,"Sagar","CSE","MP",90,),\
             (2,"Sonu","MEC","UP",60,),\
              (3,"Raghav","EEE","CH",70,),
             ]
columns=["ID","Student","Dep","City","Marks"]
df_1=spark.createDataFrame(data=SampleData_1,schema=columns)
df_1.show()
```

```
SampleData_2=[(1,"Sagar","CSE","Anmol",540,),\
             (2,"Sonu","MEC","Mahesh",450,),\
              (4,"Raghav","EEE","Jit",340,),
             ]
columns=["ID","Student","Dep","HOD","Total"]
df_2=spark.createDataFrame(data=SampleData_2,schema=columns)
df_2.show()
```

```
Final_DF=df_1.join(df_2,df_1.ID==df_2.ID,"Left")\
            .select(df_1.ID,df_1.Student,df_1.City,df_2.HOD)
Final_DF.show()
```

```

Final_DF=df_1.join(df_2,df_1.ID==df_2.ID,"inner")\
            .select(df_1.ID,df_1.Student,df_1.City,df_2.HOD)
Final_DF.show()
```

```
Final_DF=df_1.join(df_2,df_1.ID==df_2.ID,"right")\
            .select(df_1.ID,df_1.Student,df_1.City,df_2.HOD)
Final_DF.show()
```

```
Final_DF=df_1.join(df_2,df_1.ID==df_2.ID,"semi")\
            .select(df_1.ID,df_1.Student,df_1.City)
Final_DF.show()
```

```
Final_DF=df_1.join(df_2,df_1.ID==df_2.ID,"anti")\
            .select(df_1.ID,df_1.Student,df_1.City)
Final_DF.show()
```
