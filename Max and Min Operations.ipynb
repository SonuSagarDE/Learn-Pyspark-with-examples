```
import datetime
df = spark.createDataFrame([['Alex', datetime.date(1998,12,16)], ['Bob', datetime.date(1995,5,9)]], ['name', 'birthday'])
df.show()
```

```
from pyspark.sql import functions as F
col_earliest_date=F.min('birthday').alias('earliest')
col_latest_date=F.max('birthday').alias('latest')
df_result = df.select(col_earliest_date,col_latest_date)
display(df_result)
```
