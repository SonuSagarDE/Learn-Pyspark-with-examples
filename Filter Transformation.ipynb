```
SampleData_1=[(1,"Sagar","CSE","MP",90,),\
             (2,"Sonu","MEC","UP",60,),\
              (3,"Raghav","EEE","MP",70,),
              (4,"Raman","ECE","MP",65,),
             ]
columns=["ID","Student","Dep","City","Marks"]
df_1=spark.createDataFrame(data=SampleData_1,schema=columns)
df_1.show()
```

## SQL way of filtering the data

```
df_filtered=df_1.filter("City= 'MP' ")
df_filtered.show()
```

```
## Multiple Conditions
df_filtered=df_1.filter("City= 'MP'  AND Marks>=70")
df_filtered.show()
```

## Python way of filtering the data

```
df_filtered=df_1.filter(df_1["City"]=='MP')
df_filtered.show()
```

```
## Multiple Conditions
df_filtered=df_1.filter((df_1["City"]=='MP') & (df_1["Marks"]>=70)  )
df_filtered.show()
```
