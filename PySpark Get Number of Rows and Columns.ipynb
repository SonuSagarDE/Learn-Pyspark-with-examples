```
#EMP DataFrame
empData = [(1,"Smith",10,None,3000),
    (2,"Rose",20,"M",4000),
    (3,"Williams",10,"M",1000),
    (4,"Jones",10,"F",2000),
    (5,"Brown",30,"",-1),
    (6,"Brown",30,"",-1)
  ]
  
empColumns = ["emp_id","name","dept_id",
  "gender","salary"]
empDF = spark.createDataFrame(empData,empColumns)
empDF.show()
```

```
##Get Row Count
rows=empDF.count()
print(f"DataFrame row count: {rows}")
```

```
## Get Columns Count
cols=len(empDF.columns)
print(f"DataFrame columns:{cols}")
```
