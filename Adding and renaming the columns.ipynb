```
SampleData_1=[(1,"Sonu","Sagar","CSE","MP",90,),\
             (2,"Monu","Sagar","MEC","UP",60,),\
              (3,"Raghav","Juyal","EEE","MP",70,),
              (4,"Raman","Bhatia","ECE","MP",65,),
             ]
columns=["ID","FirstName","LastName","Dep","City","Marks"]
df_1=spark.createDataFrame(data=SampleData_1,schema=columns)
df_1.show()
```

```
## Add Colum of full Name
from pyspark.sql.functions import concat,col,lit
df_2=df_1.withColumn("FullName",concat(col("FirstName"),lit(' '),col("LastName")))
df_2.show()
```
